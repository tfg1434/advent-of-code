```python
from math import *
import re

with open('example.txt') as file:
    lines = file.read().splitlines()

def add(x, y):
    return f"[{x},{y}]"

"""
[[[[[9,8],1],2],3],4]

l is 5
r is 7
"""
b = 0
def explode(s, l, r):
    global b
    b += 1

    splits = []

    for i in range(l - 1, 0, -1):
        c = s[i]

        if not str.isdigit(c):
            continue

        cl = int(c) + int(s[l])

        if cl >= 10:
            splits.append(i)
            r += 1
            l += 1

        s = f"{s[:i]}{cl}{s[i+1:]}"

        break

    for i in range(r + 1, len(s)):
        c = s[i]

        if not str.isdigit(c):
            continue

        cr = int(c) \
            + int(s[r])

        if cr >= 10:
            splits.append(i)

        s = f"{s[:i]}{cr}{s[i+1:]}"

        break

    return f"{s[:l-1]}0{s[r+2:]}"#, splits

assert explode("[[[[[9,8],1],2],3],4]", 5, 7) == '[[[[0,9],2],3],4]'
assert explode("[7,[6,[5,[4,[3,2]]]]]", 13, 15) == '[7,[6,[5,[7,0]]]]'

"""
[[[[0,7],4],[15,[0,13]]],[1,1]]

l is 13
r is 14
"""
def split(s, l, r):
    i = int(s[l:r+1])
    le = floor(i / 2)
    re = ceil(i / 2)

    return f"{s[:l]}[{le},{re}]{s[r+1:]}"

assert split("[[[[0,7],4],[15,[0,13]]],[1,1]]", 13, 14) == "[[[[0,7],4],[[7,8],[0,13]]],[1,1]]"

def solve():
    for i in range(len(lines) - 1):
        lhs = lines[i]
        rhs = lines[i+1]
        res = add(lhs, rhs)

        while True:
            broke = False

            # check split
            match = re.search(r"[0-9]{2}", res) # or match?
            if match is not None:
                broke = True
                print(match.group(), match.start(), match.end() - 1)
                res = split(res, match.start(), match.end() - 1)


            # check explode
            nesting = 0
            for j, c in enumerate(res):
                if c == "[":
                    nesting += 1
                elif c == "]":
                    nesting -= 1
                
                if nesting == 5:
                    res = explode(res, j + 1, j + 3)
                    broke = True

                    break

            
            
            if not broke: break

    return res


print(solve())


```
