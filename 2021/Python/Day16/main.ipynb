```python
import numpy as np

d = {
    "0": "0000",
    "1": "0001",
    "2": "0010",
    "3": "0011",
    "4": "0100",
    "5": "0101",
    "6": "0110",
    "7": "0111",
    "8": "1000",
    "9": "1001",
    "A": "1010",
    "B": "1011",
    "C": "1100",
    "D": "1101",
    "E": "1110",
    "F": "1111",
}

with open('in.txt') as file:
    transmission = ""
    for c in file.read():
        transmission += d[c]

def parseInt(binary):
    return int(binary, 2)

vsum = 0

def solve1(packet):
    global vsum

    vsum += parseInt(packet[:3])
    packet = packet[3:]
    id = parseInt(packet[:3])
    packet = packet[3:]

    if id == 4:
        for i in range(0, len(packet), 5):
            if packet[i] == "0":
                return packet[i+5:]
    
    lId = packet[0]
    packet = packet[1:]
    
    if lId == "0":
        l = parseInt(packet[:15])
        packet = packet[15:]

        lit = packet[:l]
        packet = packet[l:]

        while lit != "":
            lit = solve1(lit)
        
        return packet

    l = parseInt(packet[:11])
    packet = packet[11:]

    for i in range(l):
        packet = solve1(packet)

    return packet

def solve2(packet):
    packet = packet[3:]
    id = parseInt(packet[:3])
    packet = packet[3:]

    if id == 4:
        b = ""

        for i in range(0, len(packet), 5):
            b += packet[i+1:i+5]

            if packet[i] == "0":
                packet = packet[i+5:]
                break

        return (packet, parseInt(b))
    
    lId = packet[0]
    packet = packet[1:]
    
    if lId == "0":
        l = parseInt(packet[:15])
        packet = packet[15:]

        lit = packet[:l]
        packet = packet[l:]

        bs = []
        while lit != "":
            lit, b = solve2(lit)
            bs.append(b)
    else:
        l = parseInt(packet[:11])
        packet = packet[11:]
        bs = []

        for i in range(l):
            packet, b = solve2(packet)
            bs.append(b)

    if id == 0: return (packet, sum(bs))
    if id == 1: return (packet, np.prod(bs))
    if id == 2: return (packet, min(bs))
    if id == 3: return (packet, max(bs))
    assert len(bs) == 2
    if id == 5: return (packet, 1 if bs[0] > bs[1] else 0)
    if id == 6: return (packet, 1 if bs[0] < bs[1] else 0)
    if id == 7: return (packet, 1 if bs[0] == bs[1] else 0)

solve1(transmission)
print(vsum)
print(solve2(transmission)[1])

```
