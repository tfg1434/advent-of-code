```python
from itertools import permutations
from functools import reduce
from numpy import clip

with open('in.txt') as file:
    lines = file.read().splitlines()

def findLows():
    ret = []

    for y in range(len(lines)):
        for x in range(len(lines[0])):
            if y != 0 and y != len(lines) - 1 and x != 0 and x != len(lines[0]) - 1:
                if lines[y - 1][x] > lines[y][x] and lines[y+1][x] > lines[y][x] and lines[y][x-1] > lines[y][x] and lines[y][x+1] > lines[y][x]:
                    ret.append([x, y])
            elif y == 0 and x == 0:
                if lines[y][x+1] > lines[y][x] and lines[y+1][x] > lines[y][x]:
                    ret.append([x, y])
            elif y == 0 and x == len(lines[0]) - 1:
                if lines[y][x-1] > lines[y][x] and lines[y+1][x] > lines[y][x]:
                    ret.append([x, y])
            elif y == len(lines) - 1 and x == 0:
                if lines[y][x+1] > lines[y][x] and lines[y-1][x] > lines[y][x]:
                    ret.append([x, y])
            elif y == len(lines) - 1 and x == len(lines[0]) - 1:
                if lines[y][x-1] > lines[y][x] and lines[y-1][x] > lines[y][x]:
                    ret.append([x, y])
            elif y == 0:
                if lines[y][x-1] > lines[y][x] and lines[y][x+1] > lines[y][x] and lines[y+1][x] > lines[y][x]:
                    ret.append([x, y])
            elif y == len(lines) - 1:
                if lines[y][x-1] > lines[y][x] and lines[y][x+1] > lines[y][x] and lines[y-1][x] > lines[y][x]:
                    ret.append([x, y])
            elif x == 0:
                if lines[y-1][x] > lines[y][x] and lines[y+1][x] > lines[y][x] and lines[y][x+1] > lines[y][x]:
                    ret.append([x, y])
            elif x == len(lines[0]) - 1:
                if lines[y-1][x] > lines[y][x] and lines[y+1][x] > lines[y][x] and lines[y][x-1] > lines[y][x]:
                    ret.append([x, y])

    return ret

def solve1():
    ans = 0

    for y in range(len(lines)):
        for x in range(len(lines[0])):
            if y != 0 and y != len(lines) - 1 and x != 0 and x != len(lines[0]) - 1:
                if lines[y - 1][x] > lines[y][x] and lines[y+1][x] > lines[y][x] and lines[y][x-1] > lines[y][x] and lines[y][x+1] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif y == 0 and x == 0:
                if lines[y][x+1] > lines[y][x] and lines[y+1][x] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif y == 0 and x == len(lines[0]) - 1:
                if lines[y][x-1] > lines[y][x] and lines[y+1][x] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif y == len(lines) - 1 and x == 0:
                if lines[y][x+1] > lines[y][x] and lines[y-1][x] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif y == len(lines) - 1 and x == len(lines[0]) - 1:
                if lines[y][x-1] > lines[y][x] and lines[y-1][x] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif y == 0:
                if lines[y][x-1] > lines[y][x] and lines[y][x+1] > lines[y][x] and lines[y+1][x] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif y == len(lines) - 1:
                if lines[y][x-1] > lines[y][x] and lines[y][x+1] > lines[y][x] and lines[y-1][x] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif x == 0:
                if lines[y-1][x] > lines[y][x] and lines[y+1][x] > lines[y][x] and lines[y][x+1] > lines[y][x]:
                    ans += 1 + int(lines[y][x])
            elif x == len(lines[0]) - 1:
                if lines[y-1][x] > lines[y][x] and lines[y+1][x] > lines[y][x] and lines[y][x-1] > lines[y][x]:
                    ans += 1 + int(lines[y][x])

    return ans

def size(low):
    s = 1

    up = lines[max(0, low[1] - 1)][low[0]]
    down = lines[min(len(lines) - 1, low[1] + 1)][low[0]]
    left = lines[low[1]][max(0, low[0] - 1)]
    right = lines[low[1]][min(len(lines[0]) - 1, low[0] + 1)]
    
    if left[0] > lines[low[1]][low[0]][0] and left[0] != "9" and left[1] != True:
        left[1] = True
        s += size((low[0] - 1, low[1]))
    
    if right[0] > lines[low[1]][low[0]][0] and right[0] != "9" and right[1] != True:
        right[1] = True
        s += size((low[0] + 1, low[1]))

    if up[0] > lines[low[1]][low[0]][0] and up[0] != "9" and up[1] != True:
        up[1] = True
        s += size((low[0], low[1] - 1))
    
    if down[0] > lines[low[1]][low[0]][0] and down[0] != "9" and down[1] != True:
        down[1] = True
        s += size((low[0], low[1] + 1))

    return s

def solve2():
    ans = []

    for low in findLows():
        ans.append(size(low))

    return reduce(lambda acc, x: acc * x, sorted(ans, reverse=True)[0:3], 1)

print(f"Part 1: {solve1()}")
lines = list(map(lambda line: list(map(lambda c: [c, False], line)), lines))
print(f"Part 2: {solve2()}")
```
