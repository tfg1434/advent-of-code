```python
with open('in.txt') as file:
    lines = file.read().splitlines()
    dots = [(int(line.split(",")[0]), int(line.split(",")[1])) for line in lines if line.find(",") != -1]
    folds = [int(line[(line.index("=") + 1):]) for line in lines if line.find("=") != -1]

def foldY(y, dots):
    dots = set(dots)
    #dots = [dot for dot in dots if dot[1] != y]
    for dot in dots.copy():
        if dot[1] == y:
            dots.remove(dot)
        if dot[1] > y:
            dots.remove(dot)
            # (dot[1] - y) + y  // - (y - dot[1])
            dy = dot[1] - y
            dots.add((dot[0], y - dy))

    return dots

def foldX(x, dots):
    dots = set(dots)
    #dots = [dot for dot in dots if dot[0] != x]
    for dot in dots.copy():
        if dot[0] == x:
            dots.remove(dot)
        if dot[0] > x:
            dots.remove(dot)
            dx = dot[0] - x
            dots.add((x - dx, dot[1]))
    
    return dots

dots = foldX(folds[0], dots)
print(len(dots))

order = ["x","y","x","y","x","y","x","y","x","y","y","y",]
for i, xy in enumerate(order):
    if xy == "x":
        dots = foldX(folds[i], dots)
    else:
        dots = foldY(folds[i], dots)

l = [[ 0 for x in range(50)] for y in range(50)]
for dot in dots:
    l[dot[0]][dot[1]] = 1

ans = ""
for y in range(50):
    for x in range(50):
        ans += "X" if (x, y) in dots else " "
    
    print(ans)
    ans = ""
```
